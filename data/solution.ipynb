```python
import pandas as pd
data = pd.read_csv("all_data.csv")
```

```python
#How many commercial chains are monitored, and therefore, included in this database?
nombreComercial =data['nombreComercial'].unique()
countComercial =pd.Series(nombreComercial).count()
print(f"The amount of commercial chains monitored are: {countComercial}")
```

```python
#What are the top 10 monitored products by State?
topByState = data.groupby(['estado','producto']).size().reset_index(name='counts').sort_values('counts',ascending = False).groupby('estado').head(10).groupby('estado')

for key, item in topByState:
    print(topByState.get_group(key))
```

```python
#Which is the commercial chain with the highest number of monitored products?
monitored = data[['nombreComercial','producto']].groupby('nombreComercial')['producto'].count().nlargest(1)
print(f"The commercial chain with the highest number of monitored product is: {monitored}")
```

```python
#Use the data to find an interesting fact.
topByState = data.groupby(['estado','producto']).size().reset_index(name='counts').sort_values('counts',ascending = True).groupby('estado').head(1).groupby('estado')
print("La menor cantidad de productos por estado:")
for key, item in topByState:
    print(topByState.get_group(key))
```
